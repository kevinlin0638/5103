```python
import math

import numpy as np
import pandas as pd

df = pd.read_csv('./data/WEP_01Nov2023_to_07Nov2023.csv')
```

```python
df_grouped = df.groupby(['PERIOD', 'sign'])['price_diff'].agg(['mean', 'count']).reset_index()
df_price = df.groupby(['PERIOD'])['USEP ($/MWh)'].agg(['mean']).reset_index()
```

```python
df_grouped
```

```python
df_price
```

```python
df_grouped['total_count'] = df_grouped.groupby(['PERIOD'])['count'].transform('sum')
df_grouped['probability'] = df_grouped['count'] / df_grouped['total_count']
df_grouped['diff_mean'] = df_grouped['mean']/100

df_grouped = df_grouped[['PERIOD', 'sign', 'diff_mean', 'probability']]
```

```python
index_col = ['PERIOD']
df_final = df_grouped.set_index(index_col).join(df_price.set_index(index_col)).reset_index()
```
